#PROJECT: Student Performance Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set(style="darkgrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df.head()
```

#Basic info

```python
print("shape",df.shape)
df.info()
```

#Checking null values

```python
df.isnull().sum()
```

#Summary


```python
df.describe()
```

#Data Visualizations

```python
df.head(5)
```

```python
sns.boxplot(x="gender",y="math score",data=df)
plt.title("Maths score performance by Gender ")
plt.show()
```


From the boxplot, we can see females perform better in reading and writing, while math scores are similar.


```python
sns.boxplot(x="test preparation course",y="writing score",data=df)
plt.title("Test performance Vs Score ")
plt.show()
```

```python
df["avg_score"]=(df["math score"]+df["reading score"]+df["writing score"])/3
df.head()
```

```python
sns.barplot(x="parental level of education",y="avg_score",data=df)
plt.title("Performance Vs Education ")
plt.show()
```

```python
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap="coolwarm")
plt.title("Score Correlation")
plt.show()
```

```python
df["Total_score"]=(df["math score"]+df["reading score"]+df["writing score"])
df.head()
```

```python
sns.lineplot(x="race/ethnicity",y="Total_score",data=df,markers="o",hue="gender")
plt.title("Trend of scores")
plt.show()
```

###  Conclusion:
- Students with test preparation scored higher on average.
- Parental education has a positive effect on scores.
- Math, reading, and writing scores are strongly correlated.
